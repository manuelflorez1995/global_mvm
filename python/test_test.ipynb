```python
import random
import pandas as pd
import numpy as np
from faker import Faker
```

```python
Fake = Faker('es_CO')
numero_dep = 4
nombre_dep = ["store_centro", "store_norte", "store_sur", "store_occidente", "store_oriente"]
departamentos = []
for i in range (1,numero_dep + 1):
    nombre = random.choice(nombre_dep)
    departamentos.append({
        "id_departamento":i,
        "nombre_departamento":f"{nombre}",
        "Ubicacion":Fake.city()
    })
departamentos_df = pd.DataFrame(departamentos)

roles = ["Analista", "Coordinador", "Director", "Asistente", "Consultor", "Supervisor", "Jefe", "Gerente"]
areas = ["Finanzas", "Recursos Humanos", "Ventas", "Marketing", "Operaciones", "Tecnología", "Datos", "Proyectos"]
niveles = ["Junior", "Semi-Senior", "Senior"]

puestos_trabajo = 3
puestos = []
for i in range (1,puestos_trabajo + 1):
    trabajo = f"{random.choice(roles)} {random.choice(niveles)} del área {random.choice(areas)}"
    puestos.append({
        "id_puesto":i,
        "titulo":trabajo,
        "salario_base":round(random.uniform(1500000, 5000000)),
        "id_departamento":random.choice(departamentos_df["id_departamento"])
    })
    
puestos_df = pd.DataFrame(puestos) 


empleados_nom = 10
empleados = []
for i in range (1,empleados_nom + 1):
    empleados.append({
        "id_empleado":i,
        "nombre":Fake.first_name(),
        "apellido":Fake.last_name(),
        "corree":Fake.email(),
        "id_puesto":random.choice(puestos_df["id_puesto"])
    })
empleados_df = pd.DataFrame(empleados)



display(departamentos_df)
display(empleados_df)
display(puestos_df)


print("Listo. Datos simulados generados en df")
```

```python
departamentos_df.to_csv('departamento.csv', index=False)
empleados_df.to_csv('empleados.csv', index=False)
puestos_df.to_csv('puestos.csv', index=False)
```
